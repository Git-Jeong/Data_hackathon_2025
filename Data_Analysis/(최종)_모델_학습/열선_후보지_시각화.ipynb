```python
import pandas as pd
```

```python
url = './make_file/(최종)_랜덤_포로스트_리스트.scv'
```

```python
data = pd.read_csv(url)
```

```python
lat_list = data['위도']
lon_list = data['경도']
```

```python
import folium

# 지도 중심 설정 (가장 첫 번째 좌표를 기준으로)
map_center = [lat_list[0], lon_list[0]]

# 지도 객체 생성
m = folium.Map(location=map_center, zoom_start=12)

# 위도, 경도 데이터를 기반으로 마커 추가
for lat, lon in zip(lat_list, lon_list):
    folium.Marker([lat, lon]).add_to(m)

# 지도 표시
m.save("./make_file/랜덤_포레스트_결과.html")

m
```
