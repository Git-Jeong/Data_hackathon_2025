```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df1 = pd.read_excel('data/사고분석-광진구(2021_2023).xlsx')
df2 = pd.read_excel('data/사고분석-광진구(2018_2020).xlsx')
df3 = pd.read_excel('data/사고분석-광진구(2015_2017).xlsx')
df4 = pd.read_excel('data/사고분석-광진구(2014).xlsx')
df = pd.concat([df1, df2, df3, df4])
df['동'] = df['시군구'].str.split(" ", expand=True)[2]
print(df['동'].value_counts())
df['발생년도'] = df['발생년월'].str.split(" ", expand=True)[0].str.replace('년', '')
df_year = df['발생년도'].value_counts().reindex(['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']).fillna(0)
print(df_year)
plt.figure(figsize=(8, 4))
plt.bar(df['동'].value_counts().index, df['동'].value_counts().values, width = 0.5)
plt.title('10년간 광진구내 동별 결빙 교통사고 누적건수')
plt.ylim(bottom = 4)
plt.grid()
plt.show()

plt.figure(figsize=(8, 4))
plt.bar(df_year.index, df_year.values, width = 0.5)
plt.title('연도별 광진구 결빙 교통사고 건수')
plt.grid()
plt.show()
```
