```python
import pandas as pd
import folium
```

```python
url = "./make_file/DBSCAN_결과.csv"
file_encoding = "UTF-8"
```

```python
data = pd.read_csv(url, encoding = file_encoding)
```

```python
data.columns
```

```python
# 기본 지도 생성 (초기 위치 설정)
map_center = [data['중점_위도'].mean(), data['중점_경도'].mean()]
m = folium.Map(location=map_center, zoom_start=12)

# 데이터의 각 행에 대해 마커 추가
for index, row in data.iterrows():
    folium.Marker([row['중점_위도'], row['중점_경도']]).add_to(m)

# 지도 출력
m.save('./make_file/DBSCAN_결과.html')
m
```
