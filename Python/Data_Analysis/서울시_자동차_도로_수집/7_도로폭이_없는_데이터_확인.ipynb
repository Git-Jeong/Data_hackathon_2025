```python
import pandas as pd
import folium
from tqdm import tqdm
```

```python
file_url = "./make_file/(최종)_서울열선_광진도로.csv"
# file_url = "./make_file/(도로폭)_서울시_도로_목록.csv"
file_encoding = "UTF-8"
```

```python
df = pd.read_csv(file_url, encoding=file_encoding, low_memory=False)
```

```python
df.columns
```

```python
print(df["도로폭"].isna().sum())
```

```python
print(df["도로규모"].isna().sum())
```

```python
data = df[df['도로 종류'].isnull() & df['도로폭'].isnull() & df['도로규모'].isnull()]
```

```python
data = df[df["도로폭"].isna() | df["도로규모"].isna()]
```

```python
start_lat = data['시작점_위도'].to_list()
start_lon = data['시작점_경도'].to_list()

end_lat = data['종료점_위도'].to_list()
end_lon = data['종료점_경도'].to_list()
```

```python
m = folium.Map(location=[sum(start_lat)/len(start_lat), sum(start_lon)/len(start_lon)], zoom_start=13)

# 선 데이터 추가
for lat1, lon1, lat2, lon2 in zip(start_lat, start_lon, end_lat, end_lon):
    folium.PolyLine([(lat1, lon1), (lat2, lon2)], color='blue', weight=2.5, opacity=1).add_to(m)

m.save('./make_file/검증/도로폭_널값_데이터_확인.html')

# 지도 출력
m
```
