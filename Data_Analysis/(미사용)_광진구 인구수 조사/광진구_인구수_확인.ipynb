### 데이터 해커톤을 대비한 주피터 노트북 코드

```python
import pandas as pd
```

```python
url = './data/서울특별시 광진구_연령별인구현황_20250203.csv'
```

```python
open_data = pd.read_csv(url, encoding = 'euc-kr')
```

```python
open_data
```

```python

# 0~9살
temp_00_09 = open_data.iloc[:2, 1:].sum(axis=0)

# 10~19살
temp_10_19 = open_data.iloc[2:4, 1:].sum(axis=0)

# 20~49살 
temp_20_49 = open_data.iloc[4:10, 1:].sum(axis=0)

# 50~59살 
temp_50_59 = open_data.iloc[10:12, 1:].sum(axis=0)

# 60~ 살
temp_60_ = open_data.iloc[12:, 1:].sum(axis=0)
 
data = pd.DataFrame({
    '9세 이하' : temp_00_09,
    '10세 이상 19세 이하' : temp_10_19,
    '20세 이상 49세 이하' : temp_20_49,
    '50세 이상 59세 이하' : temp_50_59,
    '60세 이상' : temp_60_
}) 
```

```python
data
```

```python
try :
    find = data['계']
except :
    data['계'] = data.iloc[:, :].sum(axis=1)
```

```python
try : 
    find = data['9세 이하의 비율']
except :
    data['9세 이하의 비율'] = round((data['9세 이하'] / data['계']) * 100, 2)
```

```python
try :
    find = data['19세 이하의 비율']
except :
    data['19세 이하의 비율'] = round(((data['9세 이하'] + data['10세 이상 19세 이하'] ) / data['계']) * 100, 2)
```

```python
try : 
    find = data['50세 이상의 비율']
except :
    data['50세 이상의 비율'] = round((( data['60세 이상'] + data['50세 이상 59세 이하'] ) / data['계']) * 100, 2)
```

```python
try : 
    find = data['60세 이상의 비율']
except :
    data['60세 이상의 비율'] = round( (data['60세 이상'] / data['계']) * 100, 2)
```

```python
data
```

```python
over_60 = data[['60세 이상', '60세 이상의 비율']] 
```

```python
temp = data['50세 이상 59세 이하'] + data['60세 이상']
over_50 = pd.DataFrame({
    '50세 이상' : temp,
    '50세 이상의 비율' : data['50세 이상의 비율']
})   
```
