```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import branca.colormap as cm
```

```python
# 데이터 불러오기
gwangjin = pd.read_csv('./make_file/광진구_모든_시설의_위치.csv', index_col=0)
gwangjin['시설_구분'].value_counts()
```

```python
# gwangjin['시설_구분'] 컬러맵을 범주에 맞게 설정
category_colors = {
    '어린이집': '#FFA500',  # 주황색
    '복지센터': '#1E90FF',  # 파란색
    '학교': '#32CD32',      # 초록색
    '아동센터': '#FFD700'   # 노란색
}
```

```python
import folium
from folium.plugins import HeatMap
import branca.colormap as cm  
 
# 광진구 중심 좌표 설정
gwangjin_center = [37.5374, 127.1232] 
# 기본 지도 설정 (광진구 중심)
m = folium.Map(location=gwangjin_center, zoom_start=12) 
  
# 히트맵 데이터 준비
heat_data = [[row['위도'], row['경도']] for _, row in gwangjin.iterrows()]

# 히트맵 추가
HeatMap(heat_data).add_to(m)

# 범주별 색상 맵 (Branca)
colormap = cm.linear.Set1_09.scale(0, len(gwangjin['시설_구분'].unique()))
colormap.caption = '시설 구분'
colormap.add_to(m)

# HTML로 지도 저장
m.save('./make_file/광진구_시설_분포.html')

# 마커 아이콘 
for _, row in gwangjin.iterrows(): 
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=row['시설_이름']
    ).add_to(m)

# HTML로 지도 저장
m.save('./make_file/광진구_시설_분포(마커).html')

# 지도 출력
m
```
