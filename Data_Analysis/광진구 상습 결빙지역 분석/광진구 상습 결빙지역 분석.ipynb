```python
import pandas as pd
import numpy as np
```

```python
file_url = './data/행정안전부_상습 결빙구간_20241226.csv'
```

```python
get_data = pd.read_csv(file_url, encoding = 'euc-kr')
get_data.info()
```

```python
get_data.head()
```

```python
data = get_data[get_data['관리청'] == '서울특별시(광진구)']
```

```python
data.reset_index(drop=True, inplace=True)
data.index = data.index + 1  
print(f'찾아낸 파일의 길이 : {len(data)}')
```

```python
data.to_csv('./make_file/광진구_상습_결빙_지역.csv', encoding='utf-8')
```

```python
# 지도 생성 (서울 중심 좌표)
m = folium.Map(location= [37.54995, 127.08272], zoom_start=13)

# 데이터프레임의 각 구간을 지도에 추가
for _, row in data.iterrows():
    # 중심지를 광진구로 설정
    if '광진구' in row['관리청']:
        folium.PolyLine(
            locations=[
                [row['기점 위도(WGS84(4326))'], row['기점 경도(WGS84(4326))']],
                [row['종점 위도(WGS84(4326))'], row['종점 경도(WGS84(4326))']]
            ],
            color='blue',
            weight=3,
            opacity=0.7,
            tooltip=(
                f"도로명: {row['도로(노선)명']}<br>"
                f"총길이: {row['총길이(km)']} km<br>"
                f"방위각: {row['정북기준 도로방향 각도(기점 기준십진수 방위각)']}°"
            )
        ).add_to(m)

# 지도 객체를 HTML 파일로 저장
m.save('./make_file/광진구_상습_결빙_지역.html')

# 지도 출력
m
```
