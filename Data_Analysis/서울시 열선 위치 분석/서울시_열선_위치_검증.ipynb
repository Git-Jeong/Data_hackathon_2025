```python
import pandas as pd
import folium
```

```python
url = "./make_file/열선_위경도값(중앙값_열선_경사각_포함).csv"
```

```python
data = pd.read_csv(url, encoding = "EUC-KR")
```

```python
data.columns
```

```python
start_lat = data['시점_위도'].to_list()
start_lon = data['시점_경도'].to_list()

end_lat = data['종점_위도'].to_list()
end_lon = data['종점_경도'].to_list()

avg_lat = data['중점_위도'].to_list()
avg_lon = data['중점_경도'].to_list()
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd
import folium
from shapely.geometry import LineString
```

```python
m = folium.Map(location=[sum(start_lat)/len(start_lat), sum(start_lon)/len(start_lon)], zoom_start=13)

# 선 데이터 추가
for lat1, lon1, lat2, lon2 in zip(start_lat, start_lon, end_lat, end_lon):
    folium.PolyLine([(lat1, lon1), (lat2, lon2)], color='red', weight=2.5, opacity=1).add_to(m)

m.save('./make_file/서울시_열선_시각화.html')

# 지도 출력
m
```

```python
m = folium.Map(location=[sum(start_lat)/len(start_lat), sum(start_lon)/len(start_lon)], zoom_start=13)

# 선 데이터 추가
for lat1, lon1, lat2, lon2 in zip(start_lat, start_lon, end_lat, end_lon):
    folium.PolyLine([(lat1, lon1), (lat2, lon2)], color='red', weight=2.5, opacity=1).add_to(m)

# ⭐ 평균 위치 추가 ⭐
for lat, lon in zip(avg_lat, avg_lon):
    folium.CircleMarker(
        location=[lat, lon],
        radius=3,
        color='yellow',
        fill=True,
        fill_color='yellow',
        fill_opacity=0.9
    ).add_to(m)

m.save('./make_file/(중앙)_서울시_열선_시각화.html') 

# 지도 출력
m
```
